```python
%%manim -ql SquareToCircle
# Needed to run 

##### prebuilt classes exist, EX)  class Rotating(mobject=None, *args, use_override=True, **kwargs)

class SquareToCircle(Scene):
    # At minimum need to make sure "Scene" capitalized is in class arg can add more depending on purpose
    
    def construct(self):
        # Use self to run methods 
        
        # Instantiate built in functions Circle(), Square(), etc.
        square = Square()
        circle = Circle()
        
        # Describe what each will look like
        circle.set_fill(PINK, opacity=0.5)
        
        # Make the shape in .mp4 and run
        self.play(Create(square))
        
        # run both circle and square
        self.play(Transform(square, circle))
        
        # How long it waits before restarting
        self.wait(1)
```

```python
%%manim -qm -v WARNING SquareToCircle
class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(PINK, opacity=0.5)  # set color and transparency

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(square))  # animate the creation of the square
        self.play(Transform(square, circle))  # interpolate the square into the circle
        self.play(FadeOut(square)) 
```

```python
%%manim Test

class Test(Scene):

  def construct(self):
    # Create title and location, buff is a buffer of 0.5 around the text in all 4 direction
    name = Text('Square and triangle').to_edge(UL, buff=0.5)
    
    # Create shape with attributes
    sq = Square(side_length=0.5, fill_color=RED, fill_opacity=0.3).to_edge(LEFT * 3)
    tri = Triangle().to_edge(DR)
    
    # What happens during run
    self.play(Write(name))
    
    # Creates boarder around shape
    self.play(DrawBorderThenFill(sq), run_time=2)
    
    self.play(Create(tri))
    self.wait()
    
    #move name to Up Right over 2 sec, then shape and move square(to_edge) and triangle
    self.play(name.animate.to_edge(UR), run_time = 2)
    self.play(sq.animate.scale(2), tri.animate.to_edge(DL), run_time=3)
```

```python
%%manim -ql Getters

class Getters(Scene):
  
  def construct(self):
    rec = Rectangle(color= WHITE, height= 2, width=2.5).to_edge(UL)
    cir = Circle().to_edge(DOWN)
   
    # Get arrow to follow(always_redraw) rectangle and circle

    arrow = always_redraw(lambda: Line(start= rec.get_bottom(), 
                                       end= cir.get_top(), buff= 0.2).add_tip())
 

    self.play(Create(VGroup(rec, cir, arrow)))
    self.wait()
    self.play(rec.animate.to_edge(UR), cir.animate.scale(0.5), run_time= 5)
  
```

```python
%%manim -ql Updaters

class Updaters(Scene):

  def construct(self):
    # creates a math like text
    num = MathTex('ln(2)')
    box = always_redraw(lambda:SurroundingRectangle(num, color= RED, fill_opacity= 0.3, fill_color=BLUE, buff=2))
    name = always_redraw(lambda: Text('A rectangle!').next_to(box, DOWN, buff= 0.25))

    # VGroup = move all vector objects inside ()
    self.play(Create(VGroup(num, box, name)))
    self.play(num.animate.shift(RIGHT * 2), run_time=2)
    self.wait()
```

```python
%%manim -ql ArcShapeIris

class ArcShapeIris(Scene):
    def construct(self):
        colors = [DARK_BROWN, BLUE_E, BLUE_D, BLUE_A, TEAL_B, GREEN_B, YELLOW_E]
        radius = [1 + rad * 0.1 for rad in range(len(colors))]

        circles_group = VGroup()

        circles_group.add(*[Circle(radius=rad, stroke_width=10, color=col)
                            for rad, col in zip(radius, colors)])
        self.add(circles_group)
```

```python
%%manim -ql ValueTrackers

class ValueTrackers(Scene):

  def construct(self):
    k = ValueTracker(0)

    # makes number start as 0.00 and count up to 2
    num = always_redraw(lambda: DecimalNumber().set_value(k.get_value()))
    
    # number fads in
    self.play(FadeIn(num))
    self.wait()
    #rate_func = moves at fixed rate if linear
    self.play(k.animate.set_value(1), run_time= 2, rate_func=linear)
```

```python
%%manim -ql PointMoving

class PointMoving(Scene):
    def construct(self):
        circle = Circle(radius=1, color=BLUE)
        dot = Dot()
        dot2 = dot.copy()
        self.add(dot)

        #self.play(GrowFromCenter(circle))
        self.play(Transform(dot, dot2))
        self.play(MoveAlongPath(dot, circle), run_time=2, rate_func=linear)
        self.play(Rotating(dot, run_time=2))
        self.wait()
```

```python
%%manim -ql Circular

class Circular(Scene):
    def construct(self):
        dot = Dot().set_color(ORANGE)
        l1 = Circle(LEFT, RIGHT)
        l2 = VMobject()
        self.add(d1, l1, l2)
        l2.add_updater(lambda x: x.become(Line(LEFT, d1.get_center()).set_color(ORANGE)))
        self.play(MoveAlongPath(d1, l1), rate_func=linear)
```

```python
class PointWithTrace(Scene):
    def construct(self):
        path = VMobject()
        dot = Dot()
        path.set_points_as_corners([dot.get_center(), dot.get_center()])
        def update_path(path):
            previous_path = path.copy()
            previous_path.add_points_as_corners([dot.get_center()])
            path.become(previous_path)
        path.add_updater(update_path)
        self.add(path, dot)
        self.play(Rotating(dot, radians=PI, about_point=RIGHT, run_time=2))
        self.wait()
        self.play(dot.animate.shift(UP))
        self.play(dot.animate.shift(LEFT))
        self.wait()
```

```python
%%manim -ql AroundAndEllipses

class AroundAndEllipses(ThreeDScene):
    def construct(self):
        self.camera.background_color = BLACK
        self.set_camera_orientation(phi= 60 * DEGREES, theta= 0* DEGREES)

        sphere = Sphere(radius= 1) 
        sphere.set_fill(YELLOW, opacity=1.0)
        ellipse = Ellipse(width = 5, height= 10, color= RED)
        sphere2 = Sphere(radius= 0.1,color= BLUE).shift(RIGHT)
        
        self.add(sphere) #shows yellow sphere 
        self.add(ellipse) #puts red line in
        self.add(ellipse, sphere2)
        self.play(MoveAlongPath(sphere2, ellipse), run_time=2, rate_func=linear) #makes blue sphere go around yellow sphere
```

```python
%%manim -ql AroundAndCircles

class AroundAndCircles(ThreeDScene):
    def construct(self):
        self.camera.background_color = BLACK
        self.set_camera_orientation(phi= 60 * DEGREES, theta= 0* DEGREES)

        circle = Circle(radius= 2.5, color= RED)
        sphere = Sphere(radius= 0.1,color= BLUE).shift(RIGHT)
        sphere2 = Sphere(radius= 1) 
        sphere2.set_color(YELLOW)
        
        self.add(sphere2)
        self.add(circle)
        self.play(MoveAlongPath(sphere, circle), run_time=1, rate_func=linear)
        self.add(circle, sphere)
        
```
